```python
import pandas as pd

sales = pd.read_csv('../data/clean/sales.csv')
prices = pd.read_parquet('../data/processed/prices_stores_CA.parquet')
```

```python
calendar = pd.read_parquet('../data/processed/calendar_subset.parquet')
```

```python
calendar.columns
```

```python
prices = prices.merge(calendar, on= "wm_yr_wk", how= 'left') 
```

```python
print(prices.columns,
prices.shape
)
prices
```

```python
print(prices["item_id"].nunique())
```

```python
prices_filtered = prices[prices["item_id"].isin(sales["item_id"])]
prices_filtered.shape
```

```python
prices_filtered
```

```python
weekly_price = prices_filtered.groupby(["item_id","wm_yr_wk"])["sell_price"].mean().reset_index()
```

```python
weekly_price.to_csv('../data/clean/weekly_price.csv', index=False)
```
